```python
CREATE DATABASE db_grupo_2;

--CREACIÓN DE TABLAS--

-- Tabla empleados
CREATE TABLE empleados(
id INT IDENTITY(1,1) PRIMARY KEY,
codigo VARCHAR (11) UNIQUE NOT NULL,
nombres	VARCHAR(55) UNIQUE NOT NULL,
clave VARCHAR(4) NOT NULL,
numero_documento VARCHAR(8) UNIQUE NOT NULL
);
select*from empleados;
EXEC sp_help empleados;


--Tabla administrador
CREATE TABLE administrador(
id INT IDENTITY(1,1) PRIMARY KEY,
empleado_id INT UNIQUE NOT NULL,
CONSTRAINT fk_empleados_transaccion FOREIGN KEY (empleado_id) REFERENCES empleados (id)
);
select*from administrador;
EXEC sp_help administrador;


--Tabla control
CREATE TABLE control(
id INT IDENTITY(1,1) PRIMARY KEY,
productos_dia int,
fecha_control DATE NOT NULL,
monto MONEY NOT NULL
);

select*from control;
EXEC sp_help control;


--Tabla cliente
CREATE TABLE cliente(
id INT IDENTITY(1,1) PRIMARY KEY,
nombres	VARCHAR(55) NOT NULL,
numero_documento VARCHAR(11) NOT NULL,
email VARCHAR(50) NOT NULL
pedido_id INT NOT NULL,
CONSTRAINT fk_pedido_transaccion FOREIGN KEY (pedido_id) REFERENCES pedido (id)
);
select*from cliente;
EXEC sp_help cliente;


--Tabla producto
CREATE TABLE producto (
id INT IDENTITY(1,1) PRIMARY KEY,
nombre_producto VARCHAR (50) NOT NULL,
descripcion TEXT NOT NULL,
precio_producto MONEY NOT NULL
);

select*from producto;
EXEC sp_help producto;


--Tabla combo (producto_id FK)
CREATE TABLE combo (
id INT IDENTITY(1,1) PRIMARY KEY,
nombre_combo VARCHAR (50) NOT NULL,
producto_id	INT NOT NULL,
precio_combo int  NOT NULL,
CONSTRAINT fk_producto_transaccion FOREIGN KEY (producto_id) REFERENCES producto (id)
);
ALTER TABLE combo ALTER COLUMN nombre_combo TEXT NOT NULL;

select*from combo;
EXEC sp_help combo;


--Tabla pedido
CREATE TABLE pedido (
id INT IDENTITY(1,1) PRIMARY KEY,
combo_id INT NOT NULL,
cantidad_combo int NOT NULL,
producto_id INT NOT NULL,
cantidad_producto int NOT NULL,
monto MONEY NOT NULL,
fecha_pedido DATE NOT NULL,
empleado_id INT NOT NULL,
CONSTRAINT fk_combo_id_transaccion FOREIGN KEY (combo_id) REFERENCES combo (id),
CONSTRAINT fk_producto_id_transaccion FOREIGN KEY (producto_id) REFERENCES producto (id)
CONSTRAINT fk_empleado_id_transaccion FOREIGN KEY (empleado_id) REFERENCES empleados (id);
);

select*from pedido;
EXEC sp_help pedido;

--Tabla pago
CREATE TABLE pago (
id INT IDENTITY(1,1) PRIMARY KEY,
fecha_pago DATETIME DEFAULT(GETDATE()),
pedido_id int NOT NULL
);
ALTER TABLE pago ADD CONSTRAINT fk_pedido_id_transaccion FOREIGN KEY (pedido_id) REFERENCES pedido(id);

select*from pago;
EXEC sp_help pago;
```
