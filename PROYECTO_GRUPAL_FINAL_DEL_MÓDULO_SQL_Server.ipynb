```python
Estructura del Procedimiento Almacenado “CargarFactura”


Generar 110 facturas
USE [facturacion]
EXEC CargarFactura

--------------------------
DECLARE @Counter INT = 1;

WHILE @Counter <= 100
BEGIN
	EXEC CargarFactura;
    -- Incrementar el contador
    SET @Counter = @Counter + 1;
END;

Procedimiento almacenado

ALTER PROCEDURE dbo.CargarFactura
AS
BEGIN
    SET NOCOUNT ON;

    BEGIN TRY
        -- Iniciar transacción
        BEGIN TRANSACTION;

		DECLARE @EmpleadoID INT;
		set @EmpleadoID = (select top 1 empleadoID from [dbo].[empleado] ORDER BY NEWID());
		--select @EmpleadoID

		DECLARE @ClienteID INT;
		set @ClienteID = (select top 1 [clienteID] from [dbo].[clientes] ORDER BY NEWID());
		--select @ClienteID

		DECLARE @SucursalID INT;
		set @SucursalID = (select top 1 [sucursalID] from [dbo].[sucursales] ORDER BY NEWID());
		--select @SucursalID

		--DROP TABLE #Productos;
		DECLARE @topProductos INT = (select ROUND(RAND() * 50,0) + 1);
		select top (@topProductos) * into #Productos from [dbo].[productos] ORDER BY NEWID();

		--select *from #Productos;

		---------------------------------------
        -- Insertar la factura
        DECLARE @FacturaID INT;

        INSERT INTO facturas (clienteID, sucursalID, fechaEmisionFactura, created_at, created_by)
        VALUES (@ClienteID, @SucursalID, GETDATE(), GETDATE(), @EmpleadoID);

		IF @@ROWCOUNT = 0
		BEGIN
			DECLARE @ErrorNoExiste NVARCHAR(4000) = 'No se ha insertado la factura.';
			print @ErrorNoExiste

			RAISERROR (@ErrorNoExiste, 16, 1);
		END

        SET @FacturaID = SCOPE_IDENTITY(); -- Obtener el ID de la factura recién insertada



		INSERT INTO [dbo].[informacionTributaria]
           ([facturaID]
           ,[ambiente]
           ,[tipoEmision]
           ,[razonSocial]
           ,[nombreComercial]
           ,[ruc]
           ,[claveAcceso]
           ,[codDoc]
           ,[estab]
           ,[ptoEmi]
           ,[secuencial]
           ,[dirMatriz])
     VALUES
           (@FacturaID
           ,'1' --'1: Pruebas/ 2:Producción'
           ,'01' --'Emisión Normal'
           ,'INTERAMERICANA DE PRODUCTOS QUIMICOS DEL ECUADOR S.A. PRUEBA'
           ,'PRUEBA S.A'
           ,'1790252848001'
           ,'1803202401179025222100120010020910994480000000111'
           ,'01'
           ,'001'
           ,'002'
           ,'091099448'
           ,'Km 14.5/Av. Panamericana Sur s/n Cutuglahua Casilla 17-17-021 Quito-Ecuador');



        -- Iterar sobre los productos de la factura
        DECLARE @ProductoID INT;
        DECLARE @StockActual INT;
        DECLARE @StockRestante INT;

        DECLARE product_cursor CURSOR FOR
            SELECT ProductoID FROM #Productos;

        OPEN product_cursor;

        FETCH NEXT FROM product_cursor INTO @ProductoID;

        WHILE @@FETCH_STATUS = 0
        BEGIN
			DECLARE @Cantidad INT;
			set @Cantidad = (select ROUND(RAND() * 20,0) + 1);
			--select @CantidadVenta

            -- Verificar si hay suficiente stock disponible
            SELECT @StockActual = stock FROM sucursalProductos WHERE sucursalID = @SucursalID AND productoID = @ProductoID;
            IF @StockActual >= @Cantidad
            BEGIN
                -- Actualizar el stock
                SET @StockRestante = @StockActual - @Cantidad;
                UPDATE sucursalProductos SET stock = @StockRestante WHERE sucursalID = @SucursalID AND productoID = @ProductoID;

                -- Insertar detalles de la factura
                INSERT INTO [dbo].[detallesFactura] (
					facturaID,
					productoID,
					cantidad,
					precioTotalSinImpuesto,
					subtotalDetalles)
                VALUES (
					@FacturaID,
					@ProductoID,
					@Cantidad,
					(SELECT precioUnitario FROM productos WHERE productoID = @ProductoID),
					@Cantidad * (SELECT precioUnitario FROM productos WHERE productoID = @ProductoID)
				);

                -- Incrementar el cursor
                FETCH NEXT FROM product_cursor INTO @ProductoID;
            END
            ELSE
            BEGIN
                -- Si no hay suficiente stock, hacer rollback de la transacción y salir del procedimiento con un mensaje de error
                ROLLBACK TRANSACTION;

				CLOSE product_cursor;
                DEALLOCATE product_cursor;


                THROW 51000, 'No hay suficiente stock para el producto.', 1;
                RETURN;
            END
        END

		-------------------
		INSERT INTO [dbo].[impuestoDetalle]
						   ([detallesFacturaID]
						   ,[codigo]
						   ,[codigoPorcentaje]
						   ,[tarifa]
						   ,[baseImponible]
						   ,[valor])
		SELECT [detallesFacturaID]
						   ,'2' --Código del Impuesto por producto 'IVA'
						   ,'2' --Tabla del porcentaje del IVA  codigo 2, tarifa 12%
						   ,12
						   , [subtotalDetalles]
						   , ([subtotalDetalles] * 0.12)
		FROM [dbo].[detallesFactura] df
		WHERE [facturaID] = @FacturaID;
		--------------------


		DECLARE @totalPagar MONEY = 0;
		SET @totalPagar = (SELECT SUM([subtotalDetalles]) FROM [dbo].[detallesFactura] WHERE [facturaID] = @FacturaID);

		INSERT INTO [dbo].[facturaFormaPago]
           ([facturaID]
           ,[formaPago]
           ,[total])
		VALUES
           (@FacturaID
           ,'20' --'OTROS CON UTILIZACION DEL SISTEMA FINANCIERO'
           ,@totalPagar);


		---CARGAR LOS IMPUESTOS TOTALES D ELA FACTURA

		INSERT INTO [dbo].[totalImpuesto]
           ([codigo]
           ,[codigoPorcentaje]
           ,[baseImponible]
           ,[valor]
           ,[facturaID])
		SELECT
			imp.codigo
			,imp.codigoPorcentaje
			,imp.tarifa,
			sum(imp.valor) total
			,@FacturaID
		FROM [dbo].[facturas] f INNER JOIN [dbo].[detallesFactura] df ON f.facturaID = df.facturaID
		INNER JOIN [dbo].[impuestoDetalle] imp on imp.detallesFacturaID = df.detallesFacturaID
		WHERE f.facturaID  = @FacturaID
		group by imp.codigo, imp.codigoPorcentaje, imp.tarifa;

        -- Commit de la transacción si todo fue exitoso
        COMMIT TRANSACTION;

    END TRY
    BEGIN CATCH
        -- Manejo de errores
        DECLARE @ErrorMessage NVARCHAR(4000);
        DECLARE @ErrorSeverity INT;
        DECLARE @ErrorState INT;

        SELECT
            @ErrorMessage = ERROR_MESSAGE(),
            @ErrorSeverity = ERROR_SEVERITY(),
            @ErrorState = ERROR_STATE();

        -- Rollback de la transacción en caso de error
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;

        -- Lanzar el error capturado
        RAISERROR (@ErrorMessage, @ErrorSeverity, @ErrorState);
    END CATCH;
END;


GO
```

```python
Procedimiento almacenado “SP_GENERAR_FACTURA”



USE [facturacion]

GO
/*

EXEC SP_GENERAR_FACTURA 1239

*/

go

CREATE PROCEDURE SP_GENERAR_FACTURA
@facturaID int
AS
SET NOCOUNT ON;
BEGIN

SELECT *FROM [dbo].[informacionTributaria] where [facturaID] = @facturaID;
SELECT *FROM [dbo].[facturas] where [facturaID] = @facturaID;
SELECT *FROM [dbo].[totalImpuesto] where [facturaID] = @facturaID;

SELECT *FROM [dbo].[detallesFactura] where [facturaID] = @facturaID;

SELECT *FROM [dbo].[impuestoDetalle] where [detallesFacturaID] in (
	SELECT [detallesFacturaID] FROM [dbo].[detallesFactura] where [facturaID] = @facturaID
)

SELECT *FROM [dbo].[totalImpuesto] where [facturaID] = @facturaID;
SELECT *FROM [dbo].[facturaFormaPago] where [facturaID] = @facturaID;

END

```
